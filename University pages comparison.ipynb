## University Pages Comparison


### 1. Introduction
Each university provides its own webpage by means of which it enables students and faculty to acquire information or simply stay up to date with current events. These pages may differ in many aspects. The most obvious one is the graphic layout, but the one which was of importance for this analysis was the variety of content. It can be proved quite easily by scraping the data from websites and comparing the chosen aspect.

In this short analysis, two websites have been chosen for comparison:
1. The Gdańsk University of Technology (PG) main page - [https://pg.edu.pl/en](https://pg.edu.pl/en)

2. The University of Gdańsk (UG) main page - [https://en.ug.edu.pl/](https://en.ug.edu.pl/)

### 2. Methodology
In order to show that there is a difference in content, the occurence of nouns, adjectives and verbs as well as the number of tokens were counted in both websites. This was achieved by extracting the data, parsing it, tokenizing it and finally adding the appropriate tags by using Beautiful Soup and NLTK packages. As the last step, the two results were compared by means of a simple bar chart.


#### 2.1 Data extraction
First, tha data was scraped from the PG page by means of **Beautiful Soup** package.

```python
import requests
import nltk
from bs4 import BeautifulSoup

page = requests.get("https://pg.edu.pl/en")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
paragraphs = soup.find_all('a')
text = soup.get_text().strip()
print(text)
```

In the next step, the text was tokenized and tagged using the NLTK package.

```python
lower_case = text.lower()
tokens = nltk.word_tokenize(lower_case)
tags = nltk.pos_tag(tokens)
print(tags)
```

The same was done for the UG page - with extraction as the first step, followed by tokenizing and tagging the text.

```python
page2 = requests.get("https://en.ug.edu.pl/")
```

```python
soup = BeautifulSoup(page2.content, 'html.parser')
para = soup.find_all('a')
text2 = soup.get_text().strip()
print(text2)
```

```python
lower_case2 = text2.lower()
tokens2 = nltk.word_tokenize(lower_case2)
tags2 = nltk.pos_tag(tokens2)
print(tags2)
```

```python
from collections import Counter
```

#### 2.2 Tag occurrences

The tags found in both sites' content were counted, giving the occurrence list of all parts of speech in a descending order.

The tag count for the PG page is as follows:

```python
counts = Counter( tag for word,  tag in tags)
print(counts)
```

This is the tag count for UG page:

```python
counts2 = Counter( tag for word,  tag in tags2)
print(counts2)
```

Finally, this is the token count for both sites:

```python
from nltk.probability import FreqDist

fdist = FreqDist(tokens)
print("PG:",fdist.N())

fdist2 = FreqDist(tokens2)
print("UG:",fdist2.N())
```

#### 2.3 Data representation
After collecting the data, a bar chart was created to illustrate the differences.

```python
import matplotlib.pyplot as plt

plt.bar([1,4,7],[352,85,67], color='r')
plt.bar([2,5,8],[248,67,30], color ='b')

plt.xticks([1.5,4.5,7.5],['nouns','adjectives','verbs'])
plt.grid(True)
plt.title('Nouns, adjectives and verbs count')
plt.minorticks_on()
plt.legend(labels = ('PG','UG'))
plt.show()
```

### 3. Results
As the chart shows, there is a clear difference in the occurrance of the chosen parts of speech in analysed sites. The biggest difference is in the nouns occurrences as the gap is about 100 words. There is also a difference in the number of tokens of each site, however, it is just 46 tokens, which is about 7% of both the texts' content. The gap between the tags is still relevant and can be used as a proof of the diverse nature of the webpages. Even with the tokens difference, if the whole of it was substracted from the number of nouns for instance, the gap would still remain - only it would be less distinct.
