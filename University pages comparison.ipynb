### University Pages Comparison
Each university provides its own webpage by means of which it enables students and faculty to acquire information or simply stay up to date with current events. These pages may differ in many aspects. The most obvious one is the graphic layout but the one that is of importance for this analysis is the variety of content. This can be proved quite easily by scraping the data from websites and comparing a chosen aspect.

In this short analysis, two websites have been chosen for comparison:
1. The Gdańsk University of Technology (PG) main page - [https://pg.edu.pl/en](https://pg.edu.pl/en) 
2. The University of Gdańsk (UG) main page - [https://en.ug.edu.pl/](https://en.ug.edu.pl/)


In order to show that there is a difference, the occurence of noun tags in each paragraph is counted for both websites. This is achieved by *extracting the data*, *parsing* it, *tokenizing* it and finally *adding the tags* to be counted. As the last step, the two results are compared by means of a simple bar chart.

First, tha data is scraped from the PG page by means of **Beautiful Soup**.

```python
import requests
import nltk
page = requests.get("https://pg.edu.pl/en")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
paragraphs = soup.find_all('a')
for el in paragraphs:
    text = el.get_text().strip()
    print(text)
```

In this next step, the noun occurences are counted.

```python
from collections import Counter
```

```python
noun = 0
for el in paragraphs:
    text = el.get_text().strip()
    lower_case = text.lower()
    tokens = nltk.word_tokenize(lower_case)
    tagged = nltk.pos_tag(tokens)

    counts =list(Counter(tag for word,tag in tagged))
    print(counts)
    if 'NN' in counts:
        noun += 1
    if 'NNS' in counts:
        noun += 1        
```

```python
print('Nouns in paragraphs of PG page:', noun)
```

The same is done for the UG page.

```python
page2 = requests.get("https://en.ug.edu.pl/")
```

```python
soup = BeautifulSoup(page2.content, 'html.parser')
para = soup.find_all('a')
for el in para:
    text2 = el.get_text().strip()
    print(text2)
```

```python
noun2 = 0
for el in para:
    text2 = el.get_text().strip()
    lower_case2 = text2.lower()
    tokens2 = nltk.word_tokenize(lower_case2)
    tagged2 = nltk.pos_tag(tokens2)
    
    counts =list(Counter(tag for word,tag in tagged2))
    print(counts)
    if 'NN' in counts:
        noun2 += 1
    if 'NNS' in counts:
        noun2 += 1
```

```python
print('Nouns in paragraphs of UG page:', noun2)
```

After collecting the data, a bar chart is created to illustrate the found differences.

```python
import matplotlib.pyplot as plt
```

```python
left = [1,3]
heights = [159,106]

bar_width = 0.8
plt.bar(left,heights, bar_width, color=('r','b'))

plt.xticks([0,1,2,3,4],['','PG','','UG',''])
plt.grid(True)
plt.title('Nouns in paragraphs of university pages')
plt.show()
```

As the chart shows, the PG main page contains more occurences of nouns in its parsed text's paragraphs then the UG page. It therefore proves that the two websites are different in terms of their content. Naturally, this  is just one example as further research could be performed on the already collected data - for instance by taking into account other parts of speech and calculating the percentage of their occurence on the given website. Nevertheless, the analysis already indicates that there are differences to be found by means of linguistic analysis and the right tools.
